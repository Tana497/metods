3 * x^4 - x^3 + 10 * x^2 - 5 * x - 3 = 0

### Метод ділення навпіл

```python
import numpy as np
import pandas as pd

def f(x):
    y = 3 * x**4 - x**3 + 10 * x**2 - 5 * x - 3
    return y

a = -0.4
b = 0.8
h = 0.0001

c = (a + b)/2

A = [a]
B = [b]
C = [c]
F_A = [f(a)]
F_B = [f(b)]
F_C = [f(c)]

while abs(f(c)) > h:
    if np.sign(f(c)) == np.sign(f(a)):
        a = c
    else:
        b = c
    c = (a+b)/2

    A.append(a)
    B.append(b)
    C.append(c)
    F_A.append(f(a))
    F_B.append(f(b))
    F_C.append(f(c))

print("Результати обчислень: ")
print()

data = pd.DataFrame({"a": A,
                     "F(a)": F_A,
                     "c": C,
                     "F(c)": F_C,
                     "b": B,
                    " F(b)": F_B})
print(data)

print()
print("В точці с = ", c, ", f(c) = ", f(c))
```

###Метод хорд

```python
import numpy as np
import pandas as pd

def f(x):
    y = 3 * x**4 - x**3 + 10 * x**2 - 5 * x - 3
    return y

a = -0.4
b = 0.8
h = 0.0001

d = (f(a) * b - a * f(b)) / (f(a) - f(b))

A = [a]
B = [b]
D = [d]
F_A = [f(a)]
F_B = [f(b)]
F_D = [f(d)]

while abs(f(d)) > h:
    if np.sign(f(d)) == np.sign(f(a)):
        a = d
    else:
        b = d
    d = (f(a)*b-a*f(b))/(f(a)-f(b))

    A.append(a)
    B.append(b)
    D.append(d)
    F_A.append(f(a))
    F_B.append(f(b))
    F_D.append(f(d))

print("Результати обчислень: ")
print()
data = pd.DataFrame({"a": A, "F(a)": F_A,
                     "d": D, "F(d)": F_D,
                     "b": B, "F(b)": F_B})
print(data)

print()
print("В точці с = ", d, ", f(d) = ", f(d))
```
