```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand, lambdify
```

```python
x_points = np.array([-4., -1., 1., 2.], dtype=float)
y_points = np.array([-6., 3., -11., 6.], dtype=float)
x1, x2, x3, x4 = -3, -2, -0.5, 2.5

def lagrange_polynomial(x_points, y_points):
    n = len(x_points)
    x = symbols('x')
    p = 0  # кінцевий багаточлен Лагранжа
    for i in range(n):
        # Будуємо базисний поліном l_i(x)
        l_i = 1
        for j in range(n):
            if i != j:
                l_i *= (x - x_points[j]) / (x_points[i] - x_points[j])
        p += y_points[i] * l_i
    return expand(p)  # Спрощуємо вираз багаточлена

# Отримуємо багаточлен Лагранжа
p = lagrange_polynomial(x_points, y_points)
print("Багаточлен Лагранжа:")
print(p)

# Визначаємо значення багаточлена в заданих точках
f1 = p.subs('x', x1)
print(f"Значення функції у точці x1 = {f1:.4f}")

f2 = p.subs('x', x2)
print(f"Значення функції у точці x2 = {f2:.4f}")

f3 = p.subs('x', x3)
print(f"Значення функції у точці x3 = {f3:.4f}")

f4 = p.subs('x', x4)
print(f"Значення функції у точці x4 = {f4:.4f}")
```

```python
# Побудова графіка
x_new = np.linspace(np.min(x_points), np.max(x_points), 100)
p_lambdified = lambdify(symbols('x'), p)  # Конвертуємо вираз у функцію
y_new = p_lambdified(x_new)

plt.plot(x_points, y_points, 'o', label='Задані точки')
plt.plot(x_new, y_new, label='Багаточлен Лагранжа')
plt.title('Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.interpolate import lagrange

#Перевірка
f = lagrange(x_points, y_points)
fig = plt.figure(figsize=(7, 5))
plt.plot(x_new, f(x_new), 'b', label='Lagrange Polynomial')
plt.plot(x_points, y_points, 'ro', label='Задані точки')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
